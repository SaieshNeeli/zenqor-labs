 Key Steps:

    Shuffle the data first to remove any ordering bias.

    Split into 5 equal parts (blocks).

    Randomly sample 10,000 rows from each block.

    Concatenate all sampled rows into a new DataFrame.



```python
import pandas as pd
import os

data_path = "/content/IEEE-CIS"

# Load datasets
train_transaction = pd.read_csv(os.path.join(data_path, "train_transaction.csv"))
train_identity = pd.read_csv(os.path.join(data_path, "train_identity.csv"))
test_transaction = pd.read_csv(os.path.join(data_path, "test_transaction.csv"))
test_identity = pd.read_csv(os.path.join(data_path, "test_identity.csv"))

# Merge on TransactionID
train_merged = train_transaction.merge(train_identity, how='left', on='TransactionID')
test_merged = test_transaction.merge(test_identity, how='left', on='TransactionID')

print(f" Merged Train Shape: {train_merged.shape}")
print(f" Merged Test Shape: {test_merged.shape}")

# Function to divide into 5 blocks and sample 10,000 from each
def sample_blocks(df, sample_size_per_block=10000, blocks=5, random_state=42):
    df_shuffled = df.sample(frac=1, random_state=random_state).reset_index(drop=True)  # Shuffle
    block_size = len(df_shuffled) // blocks

    samples = []
    for i in range(blocks):
        start = i * block_size
        end = (i + 1) * block_size if i < blocks - 1 else len(df_shuffled)
        block = df_shuffled.iloc[start:end]
        sample = block.sample(n=sample_size_per_block, random_state=random_state)
        samples.append(sample)

    return pd.concat(samples).reset_index(drop=True)

# Apply sampling
train_sampled = sample_blocks(train_merged)
test_sampled = sample_blocks(test_merged)

print(f" Sampled Train Shape: {train_sampled.shape}")
print(f" Sampled Test Shape: {test_sampled.shape}")
```

```python
# Step 1: Report missing value % per column in train
missing_percent = train_merged.isnull().mean().sort_values(ascending=False)
print("\n Missing Value Percentage (Train):")
print(missing_percent[missing_percent > 0])

# Step 2: Drop columns with more than 90% missing
drop_thresh = 0.9
cols_to_drop = missing_percent[missing_percent > drop_thresh].index.tolist()

train_merged.drop(columns=cols_to_drop, inplace=True)
test_merged.drop(columns=[col for col in cols_to_drop if col in test_merged.columns], inplace=True)

print(f"\n Dropped {len(cols_to_drop)} columns with >{int(drop_thresh * 100)}% missing.")

# Step 3: Add missing flags for columns with 60–90% missing
flag_thresh_low, flag_thresh_high = 0.6, 0.9
cols_with_flag = missing_percent[(missing_percent > flag_thresh_low) & (missing_percent <= flag_thresh_high)].index.tolist()

for col in cols_with_flag:
    train_merged[f"{col}_missing"] = train_merged[col].isnull().astype(int)
    if col in test_merged.columns:
        test_merged[f"{col}_missing"] = test_merged[col].isnull().astype(int)

print(f" Created {len(cols_with_flag)} missing-flag columns for features with 60–90% missing.")

# Step 4: Fill all remaining missing values with -999
train_merged.fillna(-999, inplace=True)
test_merged.fillna(-999, inplace=True)

print("\n All remaining missing values filled with -999.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
import numpy as np

# Create directories if saving plots (optional)
import os
os.makedirs("diagnostic_plots/histograms", exist_ok=True)
os.makedirs("diagnostic_plots/boxplots", exist_ok=True)

numeric_cols = train_encoded.select_dtypes(include=[np.number]).columns.difference(['TransactionID', 'isFraud'])

# 1. Initialize score dictionaries
skew_scores = {}
outlier_ratios = {}

# 2. Plot and score each numeric feature
for col in numeric_cols:
    data = train_encoded[col].copy()

    # Skip if constant
    if data.nunique() <= 1:
        continue

    # --- Calculate Skewness ---
    skew_val = skew(data)
    skew_scores[col] = abs(skew_val)

    # --- Calculate Outlier Ratio using IQR ---
    Q1, Q3 = data.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    outlier_ratio = ((data < lower) | (data > upper)).mean()
    outlier_ratios[col] = outlier_ratio

    # # --- Plot Histograms & Boxplots ---
    # plt.figure(figsize=(12, 4))

    # # Histogram
    # plt.subplot(1, 2, 1)
    # sns.histplot(data, bins=50, kde=True, color='steelblue')
    # plt.title(f"Histogram: {col} | Skew: {skew_val:.2f} | Outliers: {outlier_ratio:.2%}")

    # # Boxplot
    # plt.subplot(1, 2, 2)
    # sns.boxplot(x=data, color='orangered')
    # plt.title(f"Boxplot: {col}")

    # plt.tight_layout()
    # plt.savefig(f"diagnostic_plots/histograms/{col}_dist.png")
    # plt.savefig(f"diagnostic_plots/boxplots/{col}_box.png")
    # plt.close()  # don't display all, saves memory
```

```python
# Top skewed features
sorted_skew = sorted(skew_scores.items(), key=lambda x: x[1], reverse=True)
print("\n Top 10 Features with High Skewness:")
for col, val in sorted_skew[:10]:
    print(f"{col}: Skew = {val:.2f}")

# Top outlier-heavy features
sorted_outliers = sorted(outlier_ratios.items(), key=lambda x: x[1], reverse=True)
print("\n Top 10 Features with High Outlier Ratios:")
for col, val in sorted_outliers[:10]:
    print(f"{col}: Outlier Ratio = {val:.2%}")
```

```python
# STEP 6: Winsorization (using training-derived thresholds)

# Define thresholds based on prior scoring — filter by outlier ratio or skew score
outlier_threshold = 0.10  # ≥10% values outside IQR range
skew_threshold = 1.0      # absolute skewness ≥ 1.0

# Select features needing Winsorization
winsorize_features = [
    col for col in numeric_cols
    if outlier_ratios.get(col, 0) > outlier_threshold or skew_scores.get(col, 0) > skew_threshold
]

print(f" Winsorizing {len(winsorize_features)} features based on skew/outlier thresholds.")

winsor_limits = {}

for col in winsorize_features:
    # Compute 1st and 99th percentiles from training data only
    lower = train_encoded[col].quantile(0.01)
    upper = train_encoded[col].quantile(0.99)
    winsor_limits[col] = (lower, upper)

    # Clip train data
    train_encoded[col] = np.clip(train_encoded[col], lower, upper)

    # Clip test data using same limits — only if column exists in test
    if col in test_encoded.columns:
        test_encoded[col] = np.clip(test_encoded[col], lower, upper)

print(" Winsorization applied consistently to train and test.")
```

```python
from sklearn.preprocessing import RobustScaler

# STEP 7: Final Robust Scaler
features_to_scale = train_encoded.select_dtypes(include=[np.number]).columns.difference(['TransactionID', 'isFraud'])

# Initialize scaler
scaler = RobustScaler()

# Scale train data
train_scaled = train_encoded.copy()
train_scaled[features_to_scale] = scaler.fit_transform(train_encoded[features_to_scale])

# Scale test data using the same fitted scaler
test_scaled = test_encoded.copy()
test_scaled[features_to_scale] = scaler.transform(test_encoded[features_to_scale])

print(" Robust scaling applied to both train and test datasets.")
print(f"Final Train Shape: {train_scaled.shape}")
print(f"Final Test Shape: {test_scaled.shape}")
```

```python

# === Enhanced Anomaly Detection with Autoencoder + Isolation Forest ===

from keras.models import Model
from keras.layers import Input, Dense
from keras.optimizers import Adam
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import StandardScaler
import numpy as np

# 1. Standardize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Build Autoencoder
input_dim = X_scaled.shape[1]
encoding_dim = 32

input_layer = Input(shape=(input_dim,))
encoded = Dense(64, activation='relu')(input_layer)
encoded = Dense(encoding_dim, activation='relu')(encoded)
decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

# 3. Train Autoencoder
autoencoder.fit(X_scaled, X_scaled,
                epochs=50,
                batch_size=256,
                shuffle=True,
                validation_split=0.1,
                verbose=1)

# 4. Encode Data
encoder = Model(inputs=input_layer, outputs=encoded)
X_encoded = encoder.predict(X_scaled)

# 5. Isolation Forest on Encoded Data
iso_forest = IsolationForest(n_estimators=100, contamination=0.035, random_state=42)
iso_forest.fit(X_encoded)

# 6. Predict Anomalies
y_pred = iso_forest.predict(X_encoded)
y_scores = -iso_forest.decision_function(X_encoded)

# Map: 1 (Normal) -> 0, -1 (Anomaly) -> 1
y_pred_binary = np.where(y_pred == -1, 1, 0)

# 7. Evaluation
print("Classification Report:")
print(classification_report(y, y_pred_binary))

print("Confusion Matrix:")
print(confusion_matrix(y, y_pred_binary))

print("ROC AUC Score:")
print(roc_auc_score(y, y_scores))
```

```python
from keras.models import Model
from keras.layers import Input, Dense
from keras.optimizers import Adam
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# 1. Standardize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # <-- make sure X is clean (no NaNs)

# 2. Build Autoencoder
input_dim = X_scaled.shape[1]
encoding_dim = 32

input_layer = Input(shape=(input_dim,))
encoded = Dense(64, activation='relu')(input_layer)
encoded = Dense(encoding_dim, activation='relu')(encoded)
decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(input_dim)(decoded)  # <-- linear output (no activation)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

# 3. Train Autoencoder
history = autoencoder.fit(
    X_scaled, X_scaled,
    epochs=50,
    batch_size=256,
    shuffle=True,
    validation_split=0.1,
    verbose=1
)

# 4. Plot Loss Curve (Optional)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Autoencoder Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.legend()
plt.grid(True)
plt.show()

# 5. Encode Data
encoder = Model(inputs=input_layer, outputs=encoded)
X_encoded = encoder.predict(X_scaled)

# 6. Train Isolation Forest on Encoded Data
iso_forest = IsolationForest(n_estimators=100, contamination=0.035, random_state=42)
iso_forest.fit(X_encoded)

# 7. Predict Anomalies
y_pred = iso_forest.predict(X_encoded)
y_scores = -iso_forest.decision_function(X_encoded)

# Convert: 1 (normal) -> 0, -1 (anomaly) -> 1
y_pred_binary = np.where(y_pred == -1, 1, 0)

# 8. Evaluation
print("Classification Report:")
print(classification_report(y, y_pred_binary))

print("Confusion Matrix:")
print(confusion_matrix(y, y_pred_binary))

print("ROC AUC Score:")
print(roc_auc_score(y, y_scores))
```

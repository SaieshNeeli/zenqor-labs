```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
import numpy as np
import os
import pandas as pd

numeric_cols_train = X_train.select_dtypes(include=[np.number]).columns.tolist()


skew_scores_train = {}
outlier_ratios_train = {}

# Diagnostic loop for training data
print("\n Generating diagnostic plots and calculating metrics for numeric features in X_train...")
for col in numeric_cols_train:
    data_train = X_train[col].copy()

    # Skip columns with only one unique value (constant) in X_train
    if data_train.nunique() <= 1:
        print(f" Skipping '{col}' in X_train: only one unique value.")
        continue

    # Skew score for X_train
    # Ensure there are enough data points and variation to calculate skew
    if len(data_train) > 1 and data_train.std() > 0:
        skew_val_train = skew(data_train)
        skew_scores_train[col] = abs(skew_val_train)
    else:
        skew_val_train = 0 # Assign a skew of 0 if calculation is not possible
        skew_scores_train[col] = 0
        print(f" Skipping skew calculation for '{col}' in X_train: not enough data or variation.")


    # IQR-based outlier ratio for X_train
    Q1_train, Q3_train = data_train.quantile([0.25, 0.75])
    IQR_train = Q3_train - Q1_train

    # Handle cases where IQR is zero (e.g., all values are the same)
    if IQR_train == 0:
        outlier_ratio_train = 0
    else:
        lower_train, upper_train = Q1_train - 1.5 * IQR_train, Q3_train + 1.5 * IQR_train
        outlier_ratio_train = ((data_train < lower_train) | (data_train > upper_train)).mean()

    outlier_ratios_train[col] = outlier_ratio_train

    # --- Print summary for the current feature ---
    print(f"\nFeature (Train): {col}")
    print(f"  Skewness: {skew_val_train:.2f}")
    print(f"  Outlier Ratio: {outlier_ratio_train:.2%}")


    # --- Display plots ---
    plt.figure(figsize=(12, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data_train, bins=50, kde=True, color='steelblue')
    plt.title(f"Histogram (Train): {col}")

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data_train, color='orangered')
    plt.title(f"Boxplot (Train): {col}")

    plt.tight_layout()
    plt.show() # This will display the plot in the Colab output

print("\n Diagnostic plots and metrics generated for X_train.")
```

```python
if not skew_scores_train or not outlier_ratios_train:
    print("\n Skew scores or outlier ratios for X_train are not available.")
    print(" Please run the diagnostic code for X_train first.")
else:
    # Top skewed features from X_train
    # Use the _train versions of the dictionaries
    sorted_skew_train = sorted(skew_scores_train.items(), key=lambda x: x[1], reverse=True)
    print("\n Top 10 Features in X_train with High Skewness:")
    if sorted_skew_train:
        for col, val in sorted_skew_train[:10]:
            print(f"- {col}: Skew = {val:.2f}")
    else:
        print(" No features with calculated skew scores in X_train.")


    # Top outlier-heavy features from X_train
    # Use the _train versions of the dictionaries
    sorted_outliers_train = sorted(outlier_ratios_train.items(), key=lambda x: x[1], reverse=True)
    print("\n Top 10 Features in X_train with High Outlier Ratios:")
    if sorted_outliers_train:
        for col, val in sorted_outliers_train[:10]:
            print(f"- {col}: Outlier Ratio = {val:.2%}")
    else:
        print(" No features with calculated outlier ratios in X_train.")

print("\n End of Skewness and Outlier Summary for X_train.")
```

```python
# Define thresholds (can be adjusted based on your analysis of X_train)
outlier_threshold = 0.10   # 10% of values outside IQR range in X_train
skew_threshold = 1.0       # Skewness > 1.0 (absolute) in X_train

# Flag features for Winsorization based on metrics from X_train
winsorize_features_train = []

# Check if the dictionaries are populated before iterating
if not skew_scores_train or not outlier_ratios_train:
    print("\n Skew scores or outlier ratios for X_train are not available.")
    print(" Cannot identify features for Winsorization.")
else:
    # Get the list of numeric columns that were analyzed (those in the score dictionaries)
    analyzed_numeric_cols = set(skew_scores_train.keys()).union(outlier_ratios_train.keys())

    for col in analyzed_numeric_cols:
        # Use .get() with a default of 0 in case a feature only has a skew score or outlier ratio
        if outlier_ratios_train.get(col, 0) > outlier_threshold or skew_scores_train.get(col, 0) > skew_threshold:
            winsorize_features_train.append(col)

    # Display summary
    print(f"\n Winsorization will be considered for {len(winsorize_features_train)} feature(s) based on X_train metrics:\n")
    if winsorize_features_train:
        for col in winsorize_features_train:
            # Use .get() to safely access scores, defaulting to N/A if not found
            skew_val = skew_scores_train.get(col, 'N/A')
            outlier_val = outlier_ratios_train.get(col, 'N/A')
            print(f"- {col}: Skew = {skew_val if isinstance(skew_val, str) else f'{skew_val:.2f}'}, Outlier Ratio = {outlier_val if isinstance(outlier_val, str) else f'{outlier_val:.2%}'}")
    else:
        print(" No features in X_train meet the criteria for Winsorization.")
```

```python
# Define the percentiles for Winsorization (can be adjusted)
lower_percentile = 0.01  # 1st percentile
upper_percentile = 0.99  # 99th percentile

winsor_limits_train = {}  # Store limits calculated from X_train

print(f"\n Applying Winsorization at {lower_percentile*100:.0f}th and {upper_percentile*100:.0f}th percentiles based on X_train...")

if not winsorize_features_train:
    print(" No features flagged for Winsorization based on X_train analysis.")
else:
    for col in winsorize_features_train:
        # Check if the column exists in X_train and X_test before processing
        if col in X_train.columns and col in X_test.columns:
            # Calculate percentiles based *only* on X_train
            lower_limit = X_train[col].quantile(lower_percentile)
            upper_limit = X_train[col].quantile(upper_percentile)

            winsor_limits_train[col] = (lower_limit, upper_limit)

            # Apply clipping (Winsorization) to X_train using X_train limits
            X_train[col] = np.clip(X_train[col], lower_limit, upper_limit)

            # Apply clipping (Winsorization) to X_test using the *same* X_train limits
            X_test[col] = np.clip(X_test[col], lower_limit, upper_limit)

            print(f" {col}: Clipped X_train and X_test to [{lower_percentile*100:.0f}th (X_train)={lower_limit:.3f}, {upper_percentile*100:.0f}th (X_train)={upper_limit:.3f}]")
        else:
            print(f" Warning: Column '{col}' not found in both X_train and X_test. Skipping Winsorization for this column.")


print("\n Winsorization completed for selected features in both X_train and X_test.")
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np

time_col = 'Time'

if time_col in X_train.columns and time_col in X_test.columns:
    print(f"\n Converting '{time_col}' to datetime format...")
    # Use errors='coerce' to turn unparseable dates into NaT (Not a Time)
    X_train[time_col] = pd.to_datetime(X_train[time_col], errors='coerce')
    X_test[time_col] = pd.to_datetime(X_test[time_col], errors='coerce')

    print(f" Checking for NaT values in '{time_col}' after conversion:")
    print(f" X_train NaT count: {X_train[time_col].isnull().sum()}")
    print(f" X_test NaT count: {X_test[time_col].isnull().sum()}")


    # 2. Extract relevant features from the 'Time' column
    print(f"\n Extracting features from '{time_col}'...")
    X_train['Year'] = X_train[time_col].dt.year
    X_test['Year'] = X_test[time_col].dt.year

    X_train['Month'] = X_train[time_col].dt.month
    X_test['Month'] = X_test[time_col].dt.month

    X_train['Day'] = X_train[time_col].dt.day
    X_test['Day'] = X_test[time_col].dt.day

    X_train['DayOfWeek'] = X_train[time_col].dt.dayofweek # Monday=0, Sunday=6
    X_test['DayOfWeek'] = X_test[time_col].dt.dayofweek

    X_train['Hour'] = X_train[time_col].dt.hour
    X_test['Hour'] = X_test[time_col].dt.hour

    X_train['Minute'] = X_train[time_col].dt.minute
    X_test['Minute'] = X_test[time_col].dt.minute

    # Optionally, drop the original 'Time' column now that features are extracted
    X_train = X_train.drop(columns=[time_col])
    X_test = X_test.drop(columns=[time_col])
    print(f" Dropped original '{time_col}' column.")

    numeric_cols_after_time_features_train = X_train.select_dtypes(include=[np.number]).columns.tolist()
    object_cols_after_time_features_train = X_train.select_dtypes(include='object').columns.tolist()

    new_categorical_cols_to_encode = ['Month', 'DayOfWeek']
    new_numerical_cols_passthrough = [col for col in numeric_cols_after_time_features_train if col not in new_categorical_cols_to_encode]


    # Create the ColumnTransformer for both new and existing features
    preprocessor_time_features = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(handle_unknown='ignore'), new_categorical_cols_to_encode),
            ('num', 'passthrough', new_numerical_cols_passthrough + object_cols_after_time_features_train) # Include any existing object columns here too
        ],
        remainder='drop'
    )

    # Apply the preprocessor
    print("\n Applying One-Hot Encoding to extracted categorical time features...")
    X_train_processed_time = preprocessor_time_features.fit_transform(X_train)
    X_test_processed_time = preprocessor_time_features.transform(X_test)

    print(" Feature engineering from 'Time' and encoding completed.")
    print(f" Shape of X_train after processing 'Time': {X_train_processed_time.shape}")
    print(f" Shape of X_test after processing 'Time': {X_test_processed_time.shape}")


else:
    print(f" Error: '{time_col}' column not found in X_train or X_test. Cannot perform time feature engineering.")
```

```python
from sklearn.preprocessing import RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

numerical_cols_to_scale = X_train.select_dtypes(include=[np.number]).columns.tolist()

# Check if there are numerical columns to scale
if not numerical_cols_to_scale:
    print("\n No numerical columns found in X_train to apply RobustScaler.")
    # In this case, X_train and X_test are already in their final numerical format
    X_train_scaled = X_train
    X_test_scaled = X_test
    print(" X_train_scaled and X_test_scaled are set to the original X_train and X_test.")

else:
    # Create the RobustScaler transformer
    scaler = RobustScaler()

    scaler_preprocessor = ColumnTransformer(
        transformers=[
            ('robustscaler', scaler, numerical_cols_to_scale)
        ],
        remainder='passthrough'
    )

    # Apply scaling to X_train (Fit and Transform)
    print(f"\n Applying RobustScaler to {len(numerical_cols_to_scale)} numerical columns in X_train...")
    X_train_scaled = scaler_preprocessor.fit_transform(X_train)
    print(" RobustScaler fitted and applied to X_train.")

    # Apply scaling to X_test (Transform Only)
    print(f"\n Applying RobustScaler to the same columns in X_test...")
    X_test_scaled = scaler_preprocessor.transform(X_test)
    print(" RobustScaler applied to X_test using parameters from X_train.")

    try:
        # Get feature names after transformation (requires scikit-learn 0.23+)
        scaled_feature_names = scaler_preprocessor.get_feature_names_out()

        # Convert the scaled NumPy arrays back to DataFrames
        X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=scaled_feature_names, index=X_train.index)
        X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=scaled_feature_names, index=X_test.index)

        print("\n Converted scaled training and testing data back to DataFrames.")
        print(f" Final shape of X_train_scaled_df: {X_train_scaled_df.shape}")
        print(f" Final shape of X_test_scaled_df: {X_test_scaled_df.shape}")
        print("\n Sample of scaled X_train_scaled_df:")
        display(X_train_scaled_df.head())

    except AttributeError:
        print("\n Could not retrieve feature names automatically (requires scikit-learn 0.23+).")
        print(" Processed data is in NumPy arrays X_train_scaled and X_test_scaled.")
        print(f" Type of X_train_scaled: {type(X_train_scaled)}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, average_precision_score
import matplotlib.pyplot as plt
import seaborn as sns


def build_pipeline(
    numerical_features,
    categorical_features, # Include categorical features if any need handling
    scaler=RobustScaler(),
    imbalancer=None, # Can be SMOTE(), ADASYN(), or None
    classifier=LogisticRegression(random_state=42),
    handle_unknown_cat='ignore' # How to handle unknown categories in OHE
):
    """
    Builds a reusable machine learning pipeline with preprocessing steps.

    Args:
        numerical_features (list): List of column names for numerical features.
        categorical_features (list): List of column names for categorical features.
        scaler (transformer): Scaler object for numerical features (e.g., RobustScaler).
        imbalancer (sampler or None): Imbalance handling method (e.g., SMOTE, ADASYN), or None.
        classifier (estimator): The classification model.
        handle_unknown_cat (str): Strategy for handling unknown categories in OneHotEncoder.

    Returns:
        ImbPipeline: The scikit-learn/imblearn pipeline object.
    """

    # Create preprocessing steps
    preprocessing_steps = []

    if numerical_features:
        preprocessing_steps.append(('num_scaler', scaler, numerical_features))

    if categorical_features:
         # Use OneHotEncoder for categorical features
        preprocessing_steps.append(('cat_encoder', OneHotEncoder(handle_unknown=handle_unknown_cat), categorical_features))

    # Create the ColumnTransformer
    # Use remainder='passthrough' to keep columns not explicitly listed
    preprocessor = ColumnTransformer(
        transformers=preprocessing_steps,
        remainder='passthrough'
    )

    # Build the pipeline steps
    pipeline_steps = [('preprocessor', preprocessor)]
    if imbalancer:
         pipeline_steps.append(('imbalancer', imbalancer))

    pipeline_steps.append(('classifier', classifier))

    if imbalancer:
        pipeline = ImbPipeline(steps=pipeline_steps)
    else:
         pipeline = Pipeline(steps=pipeline_steps)

    return pipeline

numerical_features_for_pipeline = X_train.select_dtypes(include=[np.number]).columns.tolist()
categorical_features_for_pipeline = X_train.select_dtypes(include='object').columns.tolist()
new_categorical_time_features = ['Month', 'DayOfWeek']
numerical_features_for_pipeline = [col for col in numerical_features_for_pipeline if col not in new_categorical_time_features]
categorical_features_for_pipeline.extend(new_categorical_time_features)
categorical_features_for_pipeline = list(set(categorical_features_for_pipeline)) # Remove duplicates

print(f"\n Features identified for the pipeline:")
print(f" Numerical features: {len(numerical_features_for_pipeline)}")
print(f" Categorical features: {len(categorical_features_for_pipeline)}")
print("\n--- Building and Training Example Pipelines ---")

# Example 1: Logistic Regression with RobustScaler and ADASYN
print("\n Pipeline 1: Logistic Regression with RobustScaler and ADASYN")
pipeline_lr_adasyn = build_pipeline(
    numerical_features=numerical_features_for_pipeline,
    categorical_features=categorical_features_for_pipeline,
    scaler=RobustScaler(), # Use RobustScaler
    imbalancer=ADASYN(random_state=42), # Use ADASYN for imbalance
    classifier=LogisticRegression(random_state=42, solver='liblinear') # Use Logistic Regression

)

print(" Training Pipeline 1...")
pipeline_lr_adasyn.fit(X_train, y_train)
print(" Training complete.")

print("\n Evaluating Pipeline 1 on X_test...")
y_pred_lr_adasyn = pipeline_lr_adasyn.predict(X_test)
y_proba_lr_adasyn = pipeline_lr_adasyn.predict_proba(X_test)[:, 1]

print("\n Classification Report:")
print(classification_report(y_test, y_pred_lr_adasyn))

print("\n Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr_adasyn))

try:
    auc_lr_adasyn = roc_auc_score(y_test, y_proba_lr_adasyn)
    print(f"\n ROC AUC Score: {auc_lr_adasyn:.4f}")
except ValueError:
    print("\n Could not calculate ROC AUC. Ensure both classes are present in y_test.")

try:
    ap_lr_adasyn = average_precision_score(y_test, y_proba_lr_adasyn)
    print(f"\n Average Precision Score: {ap_lr_adasyn:.4f}")
except ValueError:
    print("\n Could not calculate Average Precision Score. Ensure both classes are present in y_test.")


# Example 2: Random Forest with RobustScaler and SMOTE
print("\n\n Pipeline 2: Random Forest with RobustScaler and SMOTE")
pipeline_rf_smote = build_pipeline(
    numerical_features=numerical_features_for_pipeline,
    categorical_features=categorical_features_for_pipeline,
    scaler=RobustScaler(),
    imbalancer=SMOTE(random_state=42), # Use SMOTE
    classifier=RandomForestClassifier(random_state=42, n_estimators=100) # Use Random Forest
)

print(" Training Pipeline 2...")
pipeline_rf_smote.fit(X_train, y_train)
print(" Training complete.")

print("\n Evaluating Pipeline 2 on X_test...")
y_pred_rf_smote = pipeline_rf_smote.predict(X_test)
y_proba_rf_smote = pipeline_rf_smote.predict_proba(X_test)[:, 1]

print("\n Classification Report:")
print(classification_report(y_test, y_pred_rf_smote))

print("\n Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf_smote))

try:
    auc_rf_smote = roc_auc_score(y_test, y_proba_rf_smote)
    print(f"\n ROC AUC Score: {auc_rf_smote:.4f}")
except ValueError:
    print("\n Could not calculate ROC AUC. Ensure both classes are present in y_test.")
try:
    ap_rf_smote = average_precision_score(y_test, y_proba_rf_smote)
    print(f"\n Average Precision Score: {ap_rf_smote:.4f}")
except ValueError:
    print("\n Could not calculate Average Precision Score. Ensure both classes are present in y_test.")

# Example 3: Logistic Regression with RobustScaler and NO Imbalance Handling
print("\n\n Pipeline 3: Logistic Regression with RobustScaler (No Imbalance Handling)")
pipeline_lr_noimb = build_pipeline(
    numerical_features=numerical_features_for_pipeline,
    categorical_features=categorical_features_for_pipeline,
    scaler=RobustScaler(),
    imbalancer=None, # No imbalancer
    classifier=LogisticRegression(random_state=42, solver='liblinear')
)

print(" Training Pipeline 3...")
pipeline_lr_noimb.fit(X_train, y_train)
print(" Training complete.")

print("\n Evaluating Pipeline 3 on X_test...")
y_pred_lr_noimb = pipeline_lr_noimb.predict(X_test)
y_proba_lr_noimb = pipeline_lr_noimb.predict_proba(X_test)[:, 1]

print("\n Classification Report:")
print(classification_report(y_test, y_pred_lr_noimb))

print("\n Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr_noimb))

try:
    auc_lr_noimb = roc_auc_score(y_test, y_proba_lr_noimb)
    print(f"\n ROC AUC Score: {auc_lr_noimb:.4f}")
except ValueError:
    print("\n Could not calculate ROC AUC. Ensure both classes are present in y_test.")

try:
    ap_lr_noimb = average_precision_score(y_test, y_proba_lr_noimb)
    print(f"\n Average Precision Score: {ap_lr_noimb:.4f}")
except ValueError:
    print("\n Could not calculate Average Precision Score. Ensure both classes are present in y_test.")

print("\n--- Example Pipeline Usage Complete ---")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE, ADASYN
# Note: LOF is an unsupervised/semi-supervised method and doesn't use imbalanced-learn's pipeline or samplers directly in the same way supervised models do.
# The preprocessing steps (scaling, encoding) are still relevant.

from sklearn.neighbors import LocalOutlierFactor # Import LOF
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, average_precision_score
import matplotlib.pyplot as plt
import seaborn as sns


def build_lof_pipeline(
    numerical_features,
    categorical_features,
    scaler=RobustScaler(),
    lof_params=None, # Dictionary for LOF parameters
    handle_unknown_cat='ignore'
):
    """
    Builds a preprocessing pipeline followed by a Local Outlier Factor model.
    Note: LOF is fit on the *preprocessed training data*.

    Args:
        numerical_features (list): List of column names for numerical features.
        categorical_features (list): List of column names for categorical features.
        scaler (transformer): Scaler object for numerical features (e.g., RobustScaler).
        lof_params (dict or None): Dictionary of parameters for LocalOutlierFactor.
                                    Must include novelty=True for prediction on new data.
                                    Defaults to {'n_neighbors': 20, 'contamination': 'auto', 'novelty': True}.
        handle_unknown_cat (str): Strategy for handling unknown categories in OneHotEncoder.

    Returns:
        Pipeline: The scikit-learn pipeline object with preprocessing and LOF.
    """

    # Create preprocessing steps
    preprocessing_steps = []

    if numerical_features:
        preprocessing_steps.append(('num_scaler', scaler, numerical_features))

    if categorical_features:
        # Use OneHotEncoder for categorical features
        preprocessing_steps.append(('cat_encoder', OneHotEncoder(handle_unknown=handle_unknown_cat), categorical_features))

    # Create the ColumnTransformer
    preprocessor = ColumnTransformer(
        transformers=preprocessing_steps,
        remainder='passthrough' # Keep columns not explicitly listed (if any)
    )

    # Default LOF parameters, ensuring novelty=True
    default_lof_params = {'n_neighbors': 20, 'contamination': 'auto', 'novelty': True}
    if lof_params is None:
        lof_params = default_lof_params
    else:
        # Ensure novelty=True is included, overriding if necessary
        if 'novelty' in lof_params and lof_params['novelty'] is not True:
             print("Warning: Overriding 'novelty' parameter to True for LOF pipeline.")
        lof_params['novelty'] = True


    # Initialize the LOF model with specified parameters
    lof_model = LocalOutlierFactor(**lof_params)

    # Build the pipeline steps
    # The pipeline will first preprocess data, then pass it to the LOF model
    pipeline_steps = [
        ('preprocessor', preprocessor),
        ('lof', lof_model) # LOF is the final step in this pipeline
    ]

    # Create the pipeline
    pipeline_lof = Pipeline(steps=pipeline_steps)

    return pipeline_lof


# Assuming X_train, y_train, X_test, y_test are already loaded and preprocessed up to time feature extraction.
# Identify numerical and categorical features from X_train *after* any feature engineering.
# This part should be consistent with how you defined these lists in previous cells.
numerical_features_for_pipeline = X_train.select_dtypes(include=np.number).columns.tolist()
categorical_features_for_pipeline = X_train.select_dtypes(include='object').columns.tolist()

# Adjust if you created new categorical features from 'Time' and want to OHE them
new_categorical_time_features_example = ['Month', 'DayOfWeek'] # Example, adjust based on your columns
for col in new_categorical_time_features_example:
    if col in numerical_features_for_pipeline:
        numerical_features_for_pipeline.remove(col)
        categorical_features_for_pipeline.append(col)

# Ensure no duplicates and sort
numerical_features_for_pipeline = sorted(list(set(numerical_features_for_pipeline)))
categorical_features_for_pipeline = sorted(list(set(categorical_features_for_pipeline)))


print(f"\n Features identified for the LOF pipeline:")
print(f" Numerical features: {len(numerical_features_for_pipeline)}")
print(f" Categorical features: {len(categorical_features_for_pipeline)}")
print("\n--- Building and Training LOF Pipeline ---")

# Example: Build and train a LOF pipeline
# You can pass custom lof_params if needed, e.g., lof_params={'n_neighbors': 30, 'contamination': 0.05, 'novelty': True}
pipeline_lof = build_lof_pipeline(
    numerical_features=numerical_features_for_pipeline,
    categorical_features=categorical_features_for_pipeline,
    scaler=RobustScaler() # Use RobustScaler for numerical features
    # lof_params are set by default or can be passed here
)

print("\n Training LOF Pipeline...")
# When training an LOF pipeline with novelty=True, you fit it on the training data (X_train).
# The LOF model learns the structure of the 'normal' data from X_train.
pipeline_lof.fit(X_train)
print(" Training complete.")


print("\n Evaluating LOF Pipeline on X_test...")

# Predict anomalies (-1 or 1) and get anomaly scores on the test set
# The pipeline's predict and decision_function methods call the LOF model's methods
test_predictions_lof_pipe = pipeline_lof.predict(X_test)
test_anomaly_scores_lof_pipe = pipeline_lof.decision_function(X_test)

# Evaluation metrics
# Map y_test labels to match LOF output (-1 for anomaly, 1 for inlier)
y_test_mapped_lof_pipe = y_test.map({0: -1, 1: 1})

print("\n Classification Report on Testing Data (LOF Pipeline, mapped labels):")
try:
    # Provide zero_division=1 to avoid warnings if a class has no predicted samples
    print(classification_report(y_test_mapped_lof_pipe, test_predictions_lof_pipe, zero_division=1))
except ValueError as e:
    print(f" Could not generate classification report on test data: {e}")

print("\n Confusion Matrix on Testing Data (LOF Pipeline, mapped labels):")
try:
     print(confusion_matrix(y_test_mapped_lof_pipe, test_predictions_lof_pipe))
except ValueError as e:
    print(f" Could not generate confusion matrix on test data: {e}")

# Calculate ROC AUC and Average Precision
if len(y_test.unique()) > 1:
    try:
        # For AUC/AP, use the original y_test (0=Fail/Anomaly, 1=Pass/Inlier)
        # and the negative of the anomaly scores (-decision_function)
        # Map y_test: 0 (Fail/Anomaly) -> 1 (Positive class for AUC)
        #                1 (Pass/Inlier)   -> 0 (Negative class for AUC)
        scores_for_auc_lof_pipe = -test_anomaly_scores_lof_pipe
        y_test_binary_anomaly_pipe = y_test.map({0: 1, 1: 0}) # Map 0 (Fail/Anomaly) to 1 (Positive class for AUC/AP)

        if len(y_test_binary_anomaly_pipe.unique()) < 2:
            print("\n Cannot calculate ROC AUC/AP: Only one class present in y_test after mapping.")
        else:
            # Calculate AUC
            auc_score_test_lof_pipe = roc_auc_score(y_test_binary_anomaly_pipe, scores_for_auc_lof_pipe)
            print(f"\n ROC AUC Score on Testing Data (LOF Pipeline, predicting Fail/Anomaly): {auc_score_test_lof_pipe:.4f}")

            # Calculate Average Precision Score
            ap_score_test_lof_pipe = average_precision_score(y_test_binary_anomaly_pipe, scores_for_auc_lof_pipe)
            print(f"\n Average Precision Score on Testing Data (LOF Pipeline, predicting Fail/Anomaly): {ap_score_test_lof_pipe:.4f}")

            # Plot ROC Curve
            fpr_lof_pipe, tpr_lof_pipe, thresholds_lof_pipe = roc_curve(y_test_binary_anomaly_pipe, scores_for_auc_lof_pipe)

            print("\n Visualizing ROC Curve on Testing Data (LOF Pipeline):")
            plt.figure(figsize=(8, 6))
            plt.plot(fpr_lof_pipe, tpr_lof_pipe, label=f'ROC curve (AUC = {auc_score_test_lof_pipe:.2f})')
            plt.plot([0, 1], [0, 1], 'k--') # Random guess line
            plt.xlabel('False Positive Rate')
            plt.ylabel('True Positive Rate')
            plt.title('ROC Curve for Anomaly Detection (LOF Pipeline - Test Data)')
            plt.legend(loc='lower right')
            plt.show()

    except Exception as e:
        print(f"\n Could not calculate metrics or plot ROC curve on test data: {e}")

else:
    print("\n Cannot calculate ROC AUC/AP on test data: Only one class present in y_test.")


print("\n LOF Pipeline evaluation complete.")
```

```python

# ========================
# Autoencoder + Isolation Forest for Anomaly Detection
# ========================

from keras.models import Model
from keras.layers import Input, Dense
from keras.optimizers import Adam
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import StandardScaler
import numpy as np

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Build Autoencoder
input_dim = X_train_scaled.shape[1]
encoding_dim = 32

input_layer = Input(shape=(input_dim,))
encoded = Dense(64, activation='relu')(input_layer)
encoded = Dense(encoding_dim, activation='relu')(encoded)

decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

# Train Autoencoder
autoencoder.fit(X_train_scaled, X_train_scaled,
                epochs=50,
                batch_size=32,
                shuffle=True,
                validation_split=0.1,
                verbose=1)

# Extract encoder
encoder = Model(inputs=input_layer, outputs=encoded)
X_train_encoded = encoder.predict(X_train_scaled)
X_test_encoded = encoder.predict(X_test_scaled)

# Train Isolation Forest on encoded features
isolation_forest = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
isolation_forest.fit(X_train_encoded)

# Predict anomalies
train_preds = isolation_forest.predict(X_train_encoded)
test_preds = isolation_forest.predict(X_test_encoded)

# Map y_train: 0 (Fail) -> -1, 1 (Pass) -> 1
y_train_mapped = y_train.map({0: -1, 1: 1})
y_test_mapped = y_test.map({0: -1, 1: 1})

# Evaluation
print("Train Classification Report:")
print(classification_report(y_train_mapped, train_preds))

print("Test Classification Report:")
print(classification_report(y_test_mapped, test_preds))

# AUC Score
train_scores = -isolation_forest.decision_function(X_train_encoded)
test_scores = -isolation_forest.decision_function(X_test_encoded)

train_auc = roc_auc_score((y_train == 0).astype(int), train_scores)
test_auc = roc_auc_score((y_test == 0).astype(int), test_scores)

print(f"Train ROC AUC: {train_auc:.4f}")
print(f"Test ROC AUC: {test_auc:.4f}")
```
